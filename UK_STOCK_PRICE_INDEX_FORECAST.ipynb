```python
import pandas as pd

def preprocess_data(df):
    """
    Preprocesses time series data according to specific date ranges.

    Parameters:
        data (pd.DataFrame): The input DataFrame containing time series data.

    Returns:
        pd.DataFrame: Processed DataFrame containing the time series data within the specified date range.
    """
    df_list = []
    earliest_date = pd.Timestamp('2013-06-03')
    end_date = pd.Timestamp('2023-06-30')

    for label, ts in df.items():
        # Filter the time series data within the specified date range
        active_range = ts.index.isin(pd.date_range(start=earliest_date, end=end_date))
        ts = ts[active_range].fillna(0.)

        # Create a temporary DataFrame with processed time series data
        tmp = pd.DataFrame({'stock_prices': ts})
        date = pd.to_datetime(tmp.index)
        tmp['hours_from_start'] = date.hour
        tmp['days_from_start'] = (date - earliest_date).days
        tmp['date'] = date.date
        tmp['INDEX'] = label
        tmp['day'] = date.day
        tmp['day_of_week'] = date.dayofweek
        tmp['month'] = date.month

        # Append the temporary DataFrame to the list
        df_list.append(tmp)

    # Concatenate all temporary DataFrames vertically into a single DataFrame
    time_df = pd.concat(df_list).reset_index(drop=True)

    # Filter the DataFrame to match the desired date range
    time_df = time_df[(time_df['days_from_start'] >=0) & (time_df['days_from_start'] < 7689)].copy()

    return time_df

# Assuming 'data1' is the input DataFrame containing time series data
# You can use the function like this:
processed_data = preprocess_data(df)
```

### Data Preparation

```python
# So, we have 7683  datapoints in total
len(processed_data)

#Each Index has 7683/3 = 2561 datapoints
#len(processed_data.query( "INDEX== 'FTSE 100' "))

#### Next we move to split our data

# 5,973 for training data
# 1500 for validation data = 3 index*500 = 1500

# So in the validation set, each consumer has one day of data
```

```python
import datetime
# Assuming max_prediction_length is an integer representing the number of days
max_prediction_length = 15
max_encoder_length = 1*20

# Calculate the training cutoff date by subtracting max_prediction_length days
training_cutoff = processed_data["days_from_start"].max() - max_prediction_length

training = TimeSeriesDataSet(
    processed_data[lambda x: x.days_from_start <= training_cutoff],
    time_idx="days_from_start",
    target="stock_prices",
    group_ids=["INDEX"],
    allow_missing_timesteps=True,
    min_encoder_length=max_encoder_length // 2,
    max_encoder_length=max_encoder_length,
    min_prediction_length=1,
    max_prediction_length=max_prediction_length,
    static_categoricals=["INDEX"],
    time_varying_known_reals=["hours_from_start",   "day_of_week","month"],
    time_varying_unknown_reals=['stock_prices'],
    target_normalizer=GroupNormalizer(
        groups=["INDEX"], transformation="softplus"
    ),
    add_relative_time_idx=True,
    add_target_scales=True,
    add_encoder_length=True,
)

validation = TimeSeriesDataSet.from_dataset(training, processed_data, predict=True, stop_randomization=True)


batch_size = 128
train_dataloader = training.to_dataloader(train=True, batch_size=batch_size, num_workers=4)
val_dataloader = validation.to_dataloader(train=False, batch_size=batch_size * 10, num_workers=4)
```

### Build TFT model

```python
early_stop_callback = EarlyStopping(monitor="val_loss", min_delta=1e-4, patience=5, verbose=True, mode="min")
lr_logger = LearningRateMonitor()
logger = TensorBoardLogger("lightning_logs")

trainer = pl.Trainer(
    max_epochs=10,
    accelerator='gpu',
    devices=1,
    enable_model_summary=True,
    gradient_clip_val=0.014597154486643149,
    callbacks=[lr_logger, early_stop_callback],
    logger=logger,
)


tft = TemporalFusionTransformer.from_dataset(
    training,
    learning_rate=0.0022891151653039,
    hidden_size=54,
    attention_head_size=3,
    dropout=0.15448385573807816,
    hidden_continuous_size=54,
    output_size=7,  # there are 7 quantiles by default: [0.02, 0.1, 0.25, 0.5, 0.75, 0.9, 0.98]
    loss=QuantileLoss(),
    log_interval=10,
    reduce_on_plateau_patience=4,
)
```

```python
trainer.fit(
    tft,
    train_dataloaders=train_dataloader,
    val_dataloaders=val_dataloader,
)
```

```python
best_model_path = trainer.checkpoint_callback.best_model_path
print(best_model_path)
best_tft = TemporalFusionTransformer.load_from_checkpoint(best_model_path)
```

## Test on new Data

```python
# Separate encoder and last data
encoder_data = processed_data[processed_data['days_from_start'] > processed_data['days_from_start'].max() - max_encoder_length]
last_data = processed_data[processed_data['days_from_start'] == processed_data['days_from_start'].max()]

# Convert date to datetime
last_data['date'] = pd.to_datetime(last_data['date'])

# Create decoder_data with repeated last day observations
decoder_data = pd.concat([
    last_data.assign(date=lambda x, i=i: x['date'] + pd.DateOffset(days=i))
    for i in range(1, max_prediction_length + 1)
], ignore_index=True)

# Calculate days from start
earliest_date = decoder_data['date'].min()
decoder_data['days_from_start'] = (decoder_data['date'] - earliest_date).dt.days
decoder_data['days_from_start'] += encoder_data['days_from_start'].max() + 1 - decoder_data['days_from_start'].min()


# Extract date components
decoder_data['month'] = decoder_data['date'].dt.month.astype(np.int64)
decoder_data['day'] = decoder_data['date'].dt.day.astype(np.int64)
decoder_data['day_of_week'] = decoder_data['date'].dt.dayofweek.astype(np.int64)

# Concatenate encoder and decoder data
new_prediction_data = pd.concat([encoder_data, decoder_data], ignore_index=True)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

#create out-of-sample predictions for FTSE 100
new_prediction_data=new_prediction_data.query(" INDEX == 'FTSE 100'")
new_raw_predictions = best_tft.predict(new_prediction_data, mode="raw", return_x=True)
best_tft.plot_prediction(new_raw_predictions.x, new_raw_predictions.output, idx=0, show_future_observed=False, ax=ax);
```
